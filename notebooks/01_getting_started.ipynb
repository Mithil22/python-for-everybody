## 2. Variables and Data Types

Variables store data. Python has several basic data types:
- **int**: Integer numbers (1, 2, 100)
- **float**: Decimal numbers (3.14, 2.5)
- **str**: Text strings ("hello")
- **bool**: True or False values

```python
# Integer
age = 25
print("Age:", age)
print("Type:", type(age))

# Float
height = 5.9
print("\nHeight:", height)
print("Type:", type(height))

# String
name = "Data Engineer"
print("\nName:", name)
print("Type:", type(name))

# Boolean
is_learning = True
print("\nIs Learning:", is_learning)
print("Type:", type(is_learning))
```

## 3. Basic Operations

Python supports standard mathematical operations

```python
# Arithmetic operations
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)  # Rounds down
print("Modulus (remainder):", a % b)
print("Exponentiation:", a ** b)  # 10 to the power of 3
```

## 4. String Operations

Strings are sequences of characters

```python
# String concatenation
first_name = "John"
last_name = "Doe"
full_name = first_name + " " + last_name
print("Full name:", full_name)

# String repetition
print("Ha" * 3)  # Prints: HaHaHa

# String length
print("Length of name:", len(full_name))

# String methods
message = "hello world"
print("Uppercase:", message.upper())
print("Capitalize:", message.capitalize())
print("Title Case:", message.title())
```

## 5. User Input

Get input from users (Note: input() always returns a string)

```python
# Getting user input
name = input("What is your name? ")
print("Hello, " + name + "!")

# Converting input to numbers
age_str = input("What is your age? ")
age = int(age_str)  # Convert string to integer
print("Next year you'll be", age + 1, "years old!")
```

## 7. Common Mistakes to Avoid

1. **Case sensitivity**: `name` and `Name` are different variables
2. **Indentation**: Python uses indentation (we'll see more later)
3. **Type errors**: Can't directly add strings and numbers
4. **Variable names**: Can't start with numbers, no spaces

```python
# Example of what NOT to do:
# age = "25"  # This is a string
# next_age = age + 1  # ERROR! Can't add string and number

# Correct way:
age = 25  # This is a number
next_age = age + 1  # Works!
print(next_age)
```
