## Section: Values and Types

### üìö What is a Value?

A **value** is one of the fundamental building blocks that a program works with. Think of values as the actual data your program manipulates.

**Examples of values we've seen:**
- `1`
- `2` 
- `"Hello, World!"`

### üè∑Ô∏è Understanding Types

Values belong to different **types** (also called data types). The type determines what kind of data it is and what operations you can perform on it.

#### Common Types:

| Value | Type | Description |
|-------|------|-------------|
| `2` | **integer** (int) | Whole numbers without decimals |
| `"Hello, World!"` | **string** (str) | Text enclosed in quotation marks |

```python
# Let's see these values in action
print(1)
print(2)
print("Hello, World!")
print(4)
```

### üî§ What Makes a String?

A **string** is called a "string" because it contains a "string of letters" (or characters) connected together.

**How to identify strings:**
- ‚úÖ They are enclosed in quotation marks: `"like this"`
- ‚úÖ Can use single quotes too: `'like this'`
- ‚úÖ Can contain letters, numbers, spaces, symbols

**Key Insight:** Both you (as the programmer) and the Python interpreter can identify strings by looking for the quotation marks around them!

```python
# Examples of strings with different quote styles
print("This is a string with double quotes")
print('This is a string with single quotes')
print("Strings can have numbers like 123 and symbols like @#$")
```

### üñ®Ô∏è The Print Statement

The `print` statement works with different types of values:

```python
# Print integers
print(2)
print(42)
print(100)

# Print strings
print("Hello, World!")
print("Python is awesome!")
```

---
### üîç Exploring Types with Python

Python has a built-in function called `type()` that tells you what type a value is!

```python
# Check the type of different values
print(type(2))
print(type("Hello, World!"))
print(type(42))
```
