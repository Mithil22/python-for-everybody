- The **label** (variable name) helps you find the box
- The **contents** (value) is what's stored inside

### ğŸ“ Assignment Statements

An **assignment statement** creates new variables and gives them values.

**Syntax:**
```python
variable_name = value
```

**Key Points:**
- The `=` sign means "assign" (not "equals"!)
- Variable name goes on the **left**
- Value goes on the **right**
- Read it as: "variable_name gets the value"

### ğŸ¯ Examples from the Book

Let's recreate the three assignment examples:

````python
# First assignment: assign a string to variable 'message'
message = 'And now for something completely different'

# Second assignment: assign integer 17 to variable 'n'
n = 17

# Third assignment: assign the approximate value of Ï€ to variable 'pi'
pi = 3.1415926535897931

# Let's see what we created
print("Message:", message)
print("N:", n)
print("Pi:", pi)
````

### ğŸ” Understanding Each Assignment

Let's break down what happened in each assignment:

#### Assignment 1: String Variable
```python
message = 'And now for something completely different'
```

- **Variable name:** `message`
- **Value type:** String (text)
- **What happened:** Created a new variable called `message` and stored a string in it

````python
# Check the type of message
print(type(message))
print(message)
````

#### Assignment 2: Integer Variable
```python
n = 17
```

- **Variable name:** `n`
- **Value type:** Integer (whole number)
- **What happened:** Created a new variable called `n` and stored the number 17 in it

````python
# Check the type of n
print(type(n))
print(n)
````

#### Assignment 3: Float Variable
```python
pi = 3.1415926535897931
```

- **Variable name:** `pi`
- **Value type:** Float (decimal number)
- **What happened:** Created a new variable called `pi` and stored the approximate value of Ï€ in it

**New type discovered!** This is a **float** (floating-point number) - a number with a decimal point.

````python
# Check the type of pi
print(type(pi))
print(pi)
````

### ğŸ”„ Variables Can Change

One powerful feature: you can **reassign** variables to new values!

````python
# Start with a value
x = 5
print("First value:", x)

# Change it to a new value
x = 10
print("New value:", x)

# You can even change the type!
x = "Now I'm a string!"
print("Changed to string:", x)
````

### ğŸ“ Variable Naming Rules

**Rules you MUST follow:**
- âœ… Can contain letters, numbers, and underscores
- âœ… Must start with a letter or underscore (not a number)
- âœ… Case-sensitive (`name` â‰  `Name`)
- âŒ Cannot use Python keywords (like `print`, `if`, `for`)
- âŒ Cannot contain spaces

**Good variable names:**
```python
name
user_age
total_count
firstName
_private_var
```

**Bad variable names:**
```python
2name         # Can't start with number
user-age      # Can't use hyphens
total count   # Can't have spaces
for           # Can't use Python keywords
```
